```python
#Choose subset to be downloaded
Download_subset = Buildings[Buildings.District == 'Barking and Dagenham']
```

```python
for index, row in Download_subset.iterrows():
    
    #Set file name
    file_name = '_'.join([row['District'], row['LSOA'], row['osm_id']]) + '.jpg'
    
    file_path = file_directory + file_name
    
    if os.path.isfile(file_path):
        pass
    else:
        #Extract lat,lon
        lon = row['lon']
        lat = row['lat']

        #Query Google StreetView with the random coordinates
        url = 'https://maps.googleapis.com/maps/api/streetview?location='+\
        str(lat) + ',' + str(lon) + '&fov=60' + '&size=' + size + 'x' + size + '&key=' + key

        # #Execute API query, saving photo
        urllib.urlretrieve(url, file_directory + file_name)
```
