```python
'''

This notebooks filters:

1) DeliveryClub data to within a Moscow square
2) OSM data to within the Moscow City limits

Files required:

'moscow/Moscow_AL4.shp'              <- Shapefile of Moscow city limits from OSM (Level 4)
'deliveryclub_features.csv'          <- DeliveryClub Data after non-geo feature generation
'gis.osm_buildings_a_free_1.shp'     <- Russia Central district OSM buildings data
'gis.osm_landuse_a_free_1.shp'       <- Russia Central district OSM land use data

'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import gmplot
from dateutil.parser import parse
import seaborn as sns
import time
sns.set(color_codes=True)
from collections import defaultdict
import csv
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from scipy.sparse import vstack, csr_matrix, lil_matrix
import string 
import scipy
import geopandas as gpd
from shapely.geometry import Point, Polygon
import folium
import json
import vincent
import operator

def return_lat(centroid):
    a = centroid.coords.xy
    return a[1][0]

def return_long(centroid):
    a = centroid.coords.xy
    return a[0][0]

def calculate_shape_max_diameter(shape):
    bounds = shape.bounds
    return ((bounds[2] - bounds[0])**2 + (bounds[3] - bounds[1])**2 )**0.5
```

# Load DeliveryClub data with features

```python
DC_original = pd.read_csv('DeliveryClub_Features.csv')
```

# Load Moscow Level 4 shapefile (shape of city)

```python
Moscow_boundaries_4 = gpd.read_file('moscow/Moscow_AL4.shp')
Moscow_boundaries_5 = gpd.read_file('moscow/Moscow_AL5.shp')
Moscow_boundaries_8 = gpd.read_file('moscow/Moscow_AL8.shp')

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3)

for axis in [ax1, ax2, ax3]:
    axis.set_facecolor('black')
    axis.grid(b='off')
    axis.set_aspect('equal')

fig.set_figheight(5)
fig.set_figwidth(15)

Moscow_boundaries_4.plot(ax=ax1, edgecolor='black')
Moscow_boundaries_5.plot(ax=ax2, edgecolor='black')
Moscow_boundaries_8.plot(ax=ax3, edgecolor='black')

ax1.set_title('Level 4 - City')
ax2.set_title('Level 5 - Regions')
ax3.set_title('Level 8 - Districts')

plt.show()
```

# Cut the original data to within the Moscow square

```python
#Take bounds of shape (bottom left and top right corner coords)
Moscow_bounds = Moscow_boundaries_4.geometry[0].bounds

DC_Moscow = DC_original.loc[(DC_original.delivery_long > Moscow_bounds[0]) &
                            (DC_original.delivery_long < Moscow_bounds[2]) &
                            (DC_original.delivery_lat > Moscow_bounds[1]) &
                            (DC_original.delivery_lat < Moscow_bounds[3])]
```

```python
#Save down new DeliveryClub file with delivery points within  
DC_Moscow.to_csv('DeliveryClub_Moscow.csv', index = False)
```

# Need to cut OSM data to within Moscow Level 4 shapefile

```python
#Buildings and landuse are the only two useful shapefiles
shape_files = [ 'gis.osm_buildings_a_free_1.shp', 'gis.osm_landuse_a_free_1.shp']   

buildings = gpd.read_file('../Delivery/central-fed-district-latest-free.shp/' + 'gis.osm_buildings_a_free_1.shp')
landuse = gpd.read_file('../Delivery/central-fed-district-latest-free.shp/' + 'gis.osm_landuse_a_free_1.shp')
```

```python
buildings['centroids'] = buildings.geometry.centroid
landuse['centroids'] = landuse.geometry.centroid
```

```python
#Lat and lons are extracted for all buildings
buildings['lat'] = buildings.centroids.apply(lambda x: return_lat(x))
buildings['long'] = buildings.centroids.apply(lambda x: return_long(x))

#Lat and lons are extracted for all shapefiles
landuse['lat'] = landuse.centroids.apply(lambda x: return_lat(x))
landuse['long'] = landuse.centroids.apply(lambda x: return_long(x))
```

```python
buildings

Moscow_buildings = buildings.loc[(buildings.long > Moscow_bounds[0]) &
                            (buildings.long < Moscow_bounds[2]) &
                            (buildings.lat > Moscow_bounds[1]) &
                            (buildings.lat < Moscow_bounds[3])]

Moscow_landuse = landuse.loc[(landuse.long > Moscow_bounds[0]) &
                            (landuse.long < Moscow_bounds[2]) &
                            (landuse.lat > Moscow_bounds[1]) &
                            (landuse.lat < Moscow_bounds[3])]

Moscow_buildings = Moscow_buildings.loc[:, Moscow_buildings.columns != 'centroids']
Moscow_landuse = Moscow_landuse.loc[:, Moscow_landuse.columns != 'centroids']
```

```python
#Smaller Moscow shapefiles saved down
Moscow_landuse.to_file(filename= 'Moscow_landuse.shp', driver='GeoJSON', encoding='utf-8') 
Moscow_buildings.to_file(filename= 'Moscow_buildings.shp', driver='GeoJSON', encoding='utf-8') 
```
