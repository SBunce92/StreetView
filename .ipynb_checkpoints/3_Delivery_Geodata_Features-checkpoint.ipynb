# Import the building geodata

The geodata below has already been split to be only that which occurs within the Moscow bounds

Trees are then created for the centroids of the polygon lists

```python
buildings = gpd.read_file('Moscow_buildings.shp')
landuse = gpd.read_file('Moscow_landuse.shp')

#Convert building locations into a KDTree
buildings_tree = scipy.spatial.cKDTree(buildings[['long', 'lat']].as_matrix())
#Convert building locations into a KDTree
landuse_tree = scipy.spatial.cKDTree(landuse[['long', 'lat']].as_matrix())
```

# A list of unique geopoints is generated

This reduces the delivery query length from >3m to around 110,000

```python
Moscow_food['delivery_string'] = Moscow_food['delivery_lat'].map(str) + ',' + Moscow_food['delivery_long'].map(str)
unique_delivery_strings = Moscow_food.delivery_string.unique()
```

# Generate the delivery_region_id feature (slow)

```python
def geostrings_to_poly_dict(poly_gdf, unique_geo_list, index_col):
    object_list = []
    for string in unique_geo_list:
        (lat, lon) = string.split(',')
        boundary_id = -999
        point = Point(float(lon), float(lat))
        for shape_index, shape in enumerate(poly_gdf.geometry):  
            if shape.contains(point):
                boundary_id = poly_gdf[index_col][shape_index]
        object_list.append(boundary_id)
    return pd.Series(pd.Series(object_list).values,index=unique_geo_list).to_dict()
    
delivery_boundary_dict = geostrings_to_poly_dict(Moscow_boundaries_8, unique_delivery_strings, 'ID')
Moscow_food['delivery_region_id'] = Moscow_food.delivery_string.map(delivery_boundary_dict)
```

# Generate the delivery_landuse_id features

The precise feature assigns -999 if the point does not match within a particular polygon, but the approximate feature assigns the point to the nearest polygon, if it doesn't match within a particular polygon

```python
def find_landuse_from_point(point, approximate):
    (lat, lon) = [float(i) for i in point.split(',')]
    point = Point((lon, lat))
    nearest_n_polygons = landuse.loc[landuse_tree.query((lon, lat), 5)[1]]
    #Set landuse_id to the closest polygon by default
    delivery_landuse_id = -999
    if approximate == True:
        delivery_landuse_id = nearest_n_polygons.osm_id.values[0]
    #Check if point is in the nearest_n, if yes, replace (reversed rows so closest shape is reviewed last)
    for row in nearest_n_polygons.iloc[::-1].iterrows():
        if point.within(row[1].geometry):
            delivery_landuse_id = row[1].osm_id
    return delivery_landuse_id

landuses_approximate = [find_landuse_from_point(i, approximate = True) for i in unique_delivery_strings]
landuses_precise = [find_landuse_from_point(i, approximate = False) for i in unique_delivery_strings]

LU_approx_dict = pd.Series(pd.Series(landuses_approximate).values,index=unique_delivery_strings).to_dict()
LU_precise_dict = pd.Series(pd.Series(landuses_precise).values,index=unique_delivery_strings).to_dict()

Moscow_food['delivery_LU_approx_id'] = Moscow_food.delivery_string.map(LU_approx_dict)
Moscow_food['delivery_LU_precise_id'] = Moscow_food.delivery_string.map(LU_precise_dict)
```

# Generate the delivery_building_id feature

This feature uses a kdtree to calculate the distance between a delivery point and a building centroid. If there is not a building centroid within 100m, it calculates the point as -999.

```python
def find_building_from_point(point):
    (lat, lon) = [float(i) for i in point.split(',')]
    point = Point((lon, lat))
    building_id = -999
    query = buildings_tree.query((lon, lat), 1)
    if query[0] < 0.001:
        building_id = buildings.loc[query[1]].osm_id
    return building_id
        

building_ids = [find_building_from_point(point) for point in unique_delivery_strings]
building_ids_dict = pd.Series(pd.Series(building_ids).values,index=unique_delivery_strings).to_dict()
Moscow_food['delivery_building_id'] = Moscow_food.delivery_string.map(building_ids_dict)
```
