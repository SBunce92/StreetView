```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import gmplot
from dateutil.parser import parse
import seaborn as sns
import time
sns.set(color_codes=True)
from collections import defaultdict
import csv
import re
import string 
import scipy
from operator import itemgetter
import geopandas as gpd
from shapely.geometry import Point, Polygon
import folium
import json
import vincent
import operator
import branca.colormap as cm

def retrieve_region_geometry(region_id):
    cut = Choropleth_shapefiles[Choro_number][Choropleth_shapefiles[Choro_number].osm_id == region_id]
    return cut.geometry.values[0]
```

# Import shapefiles and DeliveryClub data with geofeatures, and tidy indicies

```python
Moscow_boundaries_4 = gpd.read_file('moscow/Moscow_AL4.shp')
Moscow_boundaries_5 = gpd.read_file('moscow/Moscow_AL5.shp')
Moscow_boundaries_8 = gpd.read_file('moscow/Moscow_AL8.shp')
#Columns must be renamed to osm_id for consistency of matching
Moscow_boundaries_4 = Moscow_boundaries_4.rename(columns = {'ID':'osm_id', 'LOCALNAME':'address'})
Moscow_boundaries_5 = Moscow_boundaries_5.rename(columns = {'ID':'osm_id', 'LOCALNAME':'address'})
Moscow_boundaries_8 = Moscow_boundaries_8.rename(columns = {'ID':'osm_id', 'LOCALNAME':'address'})

buildings_raw = gpd.read_file('Moscow_buildings.shp')
landuse_raw = gpd.read_file('Moscow_landuse.shp')

landuse_raw = landuse_raw.rename(columns = {'name':'address'})

Delivery = pd.read_csv('Deliveryclub_Moscow_geo.csv')

Choropleth_levels = ['delivery_LU_approx_id','delivery_LU_precise_id','delivery_building_id', 'delivery_region_id']

#Standardise all indexes as integers
Moscow_boundaries_4.osm_id = Moscow_boundaries_4.osm_id.apply(lambda x: int(x))
Moscow_boundaries_5.osm_id = Moscow_boundaries_5.osm_id.apply(lambda x: int(x))
Moscow_boundaries_8.osm_id = Moscow_boundaries_8.osm_id.apply(lambda x: int(x))

buildings_raw.osm_id = buildings_raw.osm_id.apply(lambda x: int(x))
landuse_raw.osm_id = landuse_raw.osm_id.apply(lambda x: int(x))

for i in Choropleth_levels:
    Delivery[i] = Delivery[i].apply(lambda x: int(x))
```

# Create filter on main dataset

```python
#Trim landuse files and main DF if necessary


# Delivery Region for Horoshovskii
Choropleth_df = Delivery[Delivery.delivery_region_id == 445299]

# Delivery Region for Arbat
# Choropleth_df = Delivery[Delivery.delivery_region_id == 1257484]

# Filter for weekday/weekend
# Choropleth_df = Delivery[Delivery.weekday > 4]
# Choropleth_df = Delivery[Delivery.weekday <= 4]

# Filter for day/evening
# Choropleth_df = Delivery[Delivery.hour >= 18]
# Choropleth_df = Delivery[Delivery.hour < 18]

# Choropleth_df = Delivery
landuse = landuse_raw[landuse_raw.osm_id.isin(list(Choropleth_df.delivery_LU_approx_id.unique()))]
buildings = buildings_raw[buildings_raw.osm_id.isin(list(Choropleth_df.delivery_building_id.unique()))]
```

# Tidy Indicies and generate New Features within the Shapefiles 

```python
Choropleth_shapefiles = [landuse, landuse, buildings, Moscow_boundaries_8]

Data_columns = ['order_rating', u'order_star_rating', u'order_total', u'order_products_quantity', 
                   u'order_discount', u'price_per_item', u'weekday', u'hour', u'month', u'days_since_first',
                   u'android', u'ios', u'web',]
```

```python
#Generate New Features
a = Choropleth_df.groupby('delivery_region_id')['uuid'].nunique()
b = Choropleth_df.groupby('delivery_region_id')['delivery_address'].nunique()
c = Choropleth_df.groupby('delivery_region_id')['order_id'].nunique()

Moscow_boundaries_8['unique_users'] = Moscow_boundaries_8.osm_id.map(a)
Moscow_boundaries_8['unique_addresses'] = Moscow_boundaries_8.osm_id.map(b)
Moscow_boundaries_8['unique_orders'] = Moscow_boundaries_8.osm_id.map(c)

a = Choropleth_df.groupby('delivery_LU_approx_id')['uuid'].nunique()
b = Choropleth_df.groupby('delivery_LU_approx_id')['delivery_address'].nunique()
c = Choropleth_df.groupby('delivery_LU_approx_id')['order_id'].nunique()

landuse['unique_users'] = landuse.osm_id.map(a)
landuse['unique_addresses'] = landuse.osm_id.map(b)
landuse['unique_orders'] = landuse.osm_id.map(c)
landuse = landuse.fillna(0)

a = Choropleth_df.groupby('delivery_building_id')['uuid'].nunique()
c = Choropleth_df.groupby('delivery_building_id')['order_id'].nunique()
d = Choropleth_df.drop_duplicates('delivery_building_id')
dict_d = pd.Series(pd.Series(d.delivery_address).values,index=d.delivery_building_id).to_dict()

buildings['address'] = buildings.osm_id.map(dict_d)
buildings['unique_users'] = buildings.osm_id.map(a)
buildings['unique_orders'] = buildings.osm_id.map(c)
buildings = buildings.fillna(0)
```

# Generate Central Popups

```python
#Generate the overall foodtype_df to be used in all foodtype_dist_popups calls
overall_foodtype_df = pd.DataFrame(Delivery.chain_specialization.value_counts())
overall_foodtype_df.chain_specialization = overall_foodtype_df.chain_specialization/overall_foodtype_df.chain_specialization.sum()

#Generate the overall weekday histogram
overall_weekday = pd.Series(np.histogram(Delivery.weekday, bins=range(8), density = True)[0])

#Generate the overall hour histogram
overall_hour = pd.Series(np.histogram(Delivery.hour, bins=range(25), density = True)[0])

#Generate order_total overall histogram
upper_cut = 2000
overall_ordertotal = np.histogram(Delivery[Delivery.order_total < upper_cut].order_total, density = True)
```

```python
def admin_central_popups():
    return folium.Popup('Osm ID: ' + str(row[1].osm_id) + '<br>\
                     Name: ' + (row[1].address) + '<br>\
                     Unique Users: ' + str(row[1].unique_users) + '<br>\
                     Unique Orders: ' + str(row[1].unique_orders) + '<br>\
                     Unique Addresses: ' + str(row[1].unique_addresses)
                    )

def landuse_central_popups():
    return folium.Popup('Osm ID: ' + str(row[1].osm_id) + '<br>\
                     Landuse: ' + str(row[1].fclass) + '<br>\
                     Unique Users: ' + str(row[1].unique_users) + '<br>\
                     Unique Orders: ' + str(row[1].unique_orders) + '<br>\
                     Unique Addresses: ' + str(row[1].unique_addresses)
                    )

def building_central_popups():
    return folium.Popup('Osm ID: ' + str(row[1].osm_id) + '<br>\
                     Unique Users: ' + str(row[1].unique_users) + '<br>\
                     Building Type: ' + str(row[1]['type']) + '<br>\
                     Building Address: ' + str((row[1]['address'])).decode('utf-8') + '<br>\
                     Unique Orders: ' + str(row[1].unique_orders)
                    )

def foodtype_dist_popups():
    data = Choropleth_df[Choropleth_df[Choropleth_levels[Choro_number]] == row[1].osm_id]
    data = data[['chain_specialization', Choropleth_levels[Choro_number]]]
    
    region_df = pd.DataFrame(data.chain_specialization.value_counts())
    region_df.chain_specialization = region_df.chain_specialization/region_df.chain_specialization.sum()
    
    plotting_df = pd.concat([region_df, overall_foodtype_df], axis = 1)
    plotting_df.columns = [row[1].address, 'Overall Restaurant Type Distribution']
    plotting_df = plotting_df.sort_values('Overall Restaurant Type Distribution', ascending=False)
    
    group = vincent.GroupedBar(plotting_df)
    group.colors(brew='Set2')
    group.width=400
    group.height = 200
    group.legend(title='Categories')

    #Convert plot to json to be included in popup
    linejson = group.to_json()

    #Create popup and add chart to popup
    popup = folium.Popup(max_width=550)
    folium.Vega(linejson, height=250, width=550).add_to(popup)
    return popup

def weekday_popups():
    data = Choropleth_df[Choropleth_df[Choropleth_levels[Choro_number]] == row[1].osm_id]
    data = data[['weekday', Choropleth_levels[Choro_number]]]

    #Creating specific historgram series for each regions, based on plot in question
    region_weekday = pd.Series(np.histogram(data.weekday, 
                               bins=range(8), density = True)[0])

    #Creating the dataset that will be inset for plots
    plotting_df = pd.concat([region_weekday, overall_weekday], axis = 1)
    plotting_df.columns = [row[1].address, 'Overall Weekday Distribution']

    #Generate plot from above dataset
    line = vincent.Line(plotting_df, width=400, height=200)
    line.axis_titles(x='Weekday', y='Density')
    line.legend(title='Categories')

    #Convert plot to json to be included in popup
    linejson = line.to_json()

    #Create popup and add chart to popup
    popup = folium.Popup(max_width=550)
    folium.Vega(linejson, height=250, width=550).add_to(popup)
    return popup

def hour_popups():
    data = Choropleth_df[Choropleth_df[Choropleth_levels[Choro_number]] == row[1].osm_id]
    data = data[['hour', Choropleth_levels[Choro_number]]]

    #Creating specific historgram series for each regions, based on plot in question
    region_hour = pd.Series(np.histogram(data.hour, 
                               bins=range(25), density = True)[0])

    #Creating the dataset that will be inset for plots
    plotting_df = pd.concat([region_hour, overall_hour], axis = 1)
    plotting_df.columns = [row[1].address, 'Overall Hourly Distribution']

    #Generate plot from above dataset
    line = vincent.Line(plotting_df, width=400, height=200)
    line.axis_titles(x='Hour', y='Density')
    line.legend(title='Categories')

    #Convert plot to json to be included in popup
    linejson = line.to_json()

    #Create popup and add chart to popup
    popup = folium.Popup(max_width=550)
    folium.Vega(linejson, height=250, width=550).add_to(popup)
    return popup

def order_total_popups():
    data = Choropleth_df[Choropleth_df[Choropleth_levels[Choro_number]] == row[1].osm_id]
    data = data[['order_total', Choropleth_levels[Choro_number]]]
    data = data[data.order_total < upper_cut]

    #Creating specific historgram series for each regions, based on plot in question
    region_ordertotal = pd.Series(np.histogram(data.order_total, 
                               bins= overall_ordertotal[1], density = True)[0])

    #Creating the dataset that will be inset for plots
    plotting_df = pd.concat([pd.Series(overall_ordertotal[1]),
                             pd.Series(region_ordertotal),
                             pd.Series(overall_ordertotal[0])], axis = 1)

    plotting_df = plotting_df.set_index(0)
    plotting_df.columns = [row[1].address, 'Overall Cost Distribution']
    
    #Generate plot from above dataset
    line = vincent.Line(plotting_df, width=400, height=200)
    line.axis_titles(x='Order Total', y='Density')
    line.legend(title='Categories')

    #Convert plot to json to be included in popup
    linejson = line.to_json()

    #Create popup and add chart to popup
    popup = folium.Popup(max_width=550)
    folium.Vega(linejson, height=250, width=550).add_to(popup)
    return popup

def add_popup(x_offset, y_offset, color):
    if Choro_number == 3:
        x_offset = x_offset*10
        y_offset = y_offset*10
    folium.RegularPolygonMarker([row[1].geometry.centroid.y + y_offset,
                         row[1].geometry.centroid.x + x_offset],  
                         popup=p,
                         number_of_sides=4,
                         radius=5,
                         fill_color= color).add_to(m)        
```

# Group Data for Choropleth display

```python
Choropleth_levels = ['delivery_LU_approx_id','delivery_LU_precise_id','delivery_building_id', 'delivery_region_id']

Choro_number = 2
Data_column_number = 2
C_level = Choropleth_levels[Choro_number]
D_col = Data_columns[Data_column_number]

#Generate Choropleth IDs and the data column to represent (and a count column)
Grouped = Choropleth_df.groupby(C_level).mean().reset_index()
Ch_data = Grouped[[C_level, D_col]]
Ch_data = pd.concat([Ch_data, Choropleth_df.groupby(C_level).count().reset_index()['order_id']], axis = 1)
Ch_data = Ch_data.rename(columns = {'order_id':'count'})

#Filter out -999 values
Ch_data = Ch_data[Ch_data[C_level] != -999]

geometry_dictionary = pd.Series(Choropleth_shapefiles[Choro_number].geometry.values,
                                index=Choropleth_shapefiles[Choro_number].osm_id).to_dict()
Ch_data['geometry'] = Ch_data[C_level].map(geometry_dictionary)

choropleth_json = gpd.GeoDataFrame(Ch_data).to_json()
```

# Map Choropleths with Folium

```python
# Comment in only if you want count choropleths
# D_col = 'count'

Choro_dict = Ch_data[[C_level, D_col]].set_index(C_level)[D_col]

m = folium.Map(location=[55.764414, 37.647859], tiles='Stamen Toner')

linear = cm.LinearColormap(['red','black','blue'], 
                           vmin=Ch_data[D_col].quantile(.1),
                           vmax=Ch_data[D_col].quantile(.9))

linear.caption = D_col
m.add_child(linear)

folium.GeoJson(
    json.loads(choropleth_json),
    style_function=lambda feature: {
        'fillColor': linear(Choro_dict[feature['properties'][C_level]]),
        'color' : 'black',
        'weight' : 1,
        'dashArray' : '5, 5'
        },
    smooth_factor = 1.0
    ).add_to(m)

#Adding central points
for row in Choropleth_shapefiles[Choro_number].iterrows():   
    if Choro_number == 3:     
        p = admin_central_popups()      
    if Choro_number <= 1:     
        p = landuse_central_popups()       
    if Choro_number == 2:     
        p = building_central_popups()   
    add_popup(x_offset = 0, y_offset = 0, color = 'orange')

# #Adding food type distribution points    
# for row in Choropleth_shapefiles[Choro_number].iterrows():   
#     p = foodtype_dist_popups()  
#     add_popup(x_offset = 0.0001, y_offset = 0.0001, color = 'blue')  

# #Adding weekday histogram graph
# for row in Choropleth_shapefiles[Choro_number].iterrows():   
#     p = weekday_popups() 
#     add_popup(x_offset = -0.0001, y_offset = -0.0001, color = 'green')  

# #Adding hour histogram graph
# for row in Choropleth_shapefiles[Choro_number].iterrows():   
#     p = hour_popups() 
#     add_popup(x_offset = -0.0001, y_offset = 0.0001, color = 'red')  
    
# #Adding order_total histogram graph
# for row in Choropleth_shapefiles[Choro_number].iterrows():   
#     p = order_total_popups()     
#     add_popup(x_offset = 0.0001, y_offset = -0.0001, color = 'purple')  

     
file_name = './visualisations/new/' +  str(C_level) + '_' + str(D_col) + '_Arbat.html'
m.save(file_name)
```
